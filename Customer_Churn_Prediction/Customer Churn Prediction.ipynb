```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(features, labels , test_size=0.3, random_state=1)
```

```python
#from EDA we find that the data is unbalanced so we will upsampling data 
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy = 1, random_state=1)   
x_train_s, y_train_s = sm.fit_resample(x_train, y_train.values.ravel())
                                         
```

```python
print('After upsampling count of label 0 {}'.format(sum(y_train_s==0)))
print('After upsampling count of label 1 {}'.format(sum(y_train_s==1)))
```

```python
#compare between classifcation models to choose best one 
#1-random forest
#2-Support Vector classifier 
#3-Logistic Regression
```

```python
#we care the most for the false negative 
#we need to reduce the number of chrun clients that the model predict that the won't chrun 
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import cohen_kappa_score
from xgboost import XGBClassifier
```

```python
#1-Random forest 
rfc = RandomForestClassifier()
rfc.fit(x_train_s, y_train_s)
y_pred = rfc.predict(x_test)
print('Classification report: ')
print('{}'.format(classification_report(y_test, y_pred)))
print('Confusion Matrix')
print('{}'.format(confusion_matrix(y_test, y_pred)))
```

```python
#support vector classifier
svc = SVC(kernel='rbf', decision_function_shape='ovr')
svc.fit(x_train_s, y_train_s)
y_pred = svc.predict(x_test)
print('Classification report: ')
print('{}'.format(classification_report(y_test, y_pred)))
print('Confusion Matrix')
print('{}'.format(confusion_matrix(y_test, y_pred)))
```

```python
#LogisticRegression
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression()
lrc.fit(x_train_s, y_train_s)
y_pred = lrc.predict(x_test)
print('Classification report: ')
print('{}'.format(classification_report(y_test, y_pred)))
print('Confusion Matrix')
print('{}'.format(confusion_matrix(y_test, y_pred)))
```

```python
#we care about false negative and false positive 
#false negative means that the model predict that the client won't chrun but it chrun 
#so we care aboout recall 
#Here the highest value for recall find in logitistic model 
```

```python
#fine tuning the logistic regression parameters to get better model
```

```python
#Step 1 - Import the library - GridSearchCv
import numpy as np
from sklearn import linear_model, decomposition, datasets
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
logistic_Reg =LogisticRegression()
```

```python
 pipe = Pipeline(steps=[('logistic_Reg', logistic_Reg)])
```

```python
#Logistic Regression requires two parameters 'C' and 'penalty' to be optimised by GridSearchCV. So we have set these two parameters
# as a list of values form which GridSearchCV will select the best value of parameter.
C = np.logspace(-4, 4, 50)
penalty = ['l1', 'l2']
```

```python
#Now we are creating a dictionary to set all the parameters options for different modules
parameters = dict(logistic_Reg__C=C,logistic_Reg__penalty=penalty)
```

```python
#Using GridSearchCV and Printing Results
```

```python
clf_GS = GridSearchCV(pipe,parameters)
clf_GS.fit(x_train_s, y_train_s)
```

```python
print('Best Penalty:', clf_GS.best_estimator_.get_params()['logistic_Reg__penalty'])
print('Best C:', clf_GS.best_estimator_.get_params()['logistic_Reg__C'])
   
```

```python
from sklearn.linear_model import LogisticRegression
best_lrc=LogisticRegression(C=0.0009102981779915217,penalty='l2')
#LogisticRegression
#from sklearn.linear_model import LogisticRegression
best_lrc.fit(x_train_s, y_train_s)
y_pred = best_lrc.predict(x_test)
print('Classification report: ')
print('{}'.format(classification_report(y_test, y_pred)))
print('Confusion Matrix')
print('{}'.format(confusion_matrix(y_test, y_pred)))
```
